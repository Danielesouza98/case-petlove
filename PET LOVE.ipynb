```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from IPython.display import display
```

```python
df = pd.read_csv("C:/Users/Daniele/Downloads/data-test-analytics_5.csv")
```

```python
display(df)
```

# Convertendo colunas para o tipo datetime para facilitar a análise temporal.

```python
# Convertendo colunas para o tipo datetime
df['created_at'] = pd.to_datetime(df['created_at'])
df['last_date_purchase'] = pd.to_datetime(df['last_date_purchase'])
df['deleted_at'] = pd.to_datetime(df['deleted_at'])
df['birth_date'] = pd.to_datetime(df['birth_date'])
df['updated_at'] = pd.to_datetime(df['updated_at'])
df['deleted_at'] = df['deleted_at'].dt.to_period('M').dt.to_timestamp()
```

```python
display(df)
```

# TABELA CONTENDO OS CLIENTES QUE CANCELARAM O PLANO

```python
df_cancelados = df[df['status'] == 'canceled']
```

```python
display(df_cancelados)
```

```python
df_cancelados.info()
```

# Contagem de cancelamentos por mês do ano

```python
cancelamentos_por_mes = df_cancelados.groupby('deleted_at')['status'].count().reset_index()

# Renomeando a colunas para 'ano_mes' e 'total_cancelamentos'
cancelamentos_por_mes.rename(columns={'deleted_at':'ano_mes','status': 'total_cancelamentos'}, inplace=True)

# Ordena por ano_mes
cancelamentos_por_mes.sort_values(by=['ano_mes'], inplace=True)
```

```python
display(cancelamentos_por_mes)
```

```python
# Configurando o gráfico de barras
plt.figure(figsize=(15, 6))
plt.bar(cancelamentos_por_mes['ano_mes'], cancelamentos_por_mes['total_cancelamentos'], width=20, color='purple')

# Personalizando o gráfico
plt.xlabel('Ano-Mês')
plt.ylabel('Total de Cancelamentos')
plt.title('Total de Cancelamentos por Ano-Mês')

# Formatando as datas no eixo x como 'ano-mes'
date_form = mdates.DateFormatter('%Y-%m')
plt.gca().xaxis.set_major_formatter(date_form)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1))

# Rotacionando as labels do eixo x para melhor visualização
plt.xticks(rotation=45, ha='right')

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```
